# 06 — Demo in tempo reale
Rilevamento + overlay di compliance su un'immagine / cartella. Vedi anche `app/streamlit_app.py`.

````python
import sys, pathlib
root = pathlib.Path.cwd().parent
if str(root) not in sys.path: sys.path.insert(0, str(root))
from src.config import set_reproducibility, select_device, SH17_CLASSES
set_reproducibility(42); print('device:', select_device('auto'))
````

````python
import cv2, matplotlib.pyplot as plt
from src.inference import load_model, predict_image
from src.compliance import assess_compliance, summarize_compliance
from src.visualization import draw_detections, draw_compliance
from src.postprocess import filter_detections

model = load_model(root / 'models/best.pt')
img = cv2.imread('../samples/example.jpg')
dets = filter_detections(predict_image(model, img, conf=0.25), conf_threshold=0.25)
ass = assess_compliance(dets)
out = draw_compliance(draw_detections(img, dets, SH17_CLASSES), ass)
print(summarize_compliance(ass))
plt.figure(figsize=(12,8)); plt.imshow(cv2.cvtColor(out, cv2.COLOR_BGR2RGB)); plt.axis('off'); plt.show()
````

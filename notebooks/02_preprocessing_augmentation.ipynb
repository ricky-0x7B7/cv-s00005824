# 02 — Preprocessing e Augmentation
Augmentation consapevole del dominio (nessuna rotazione/ribaltamento verticale). Visualizzazione prima/dopo.

```python
import sys, pathlib
root = pathlib.Path.cwd().parent
if str(root) not in sys.path: sys.path.insert(0, str(root))
from src.config import set_reproducibility, select_device, SH17_CLASSES
set_reproducibility(42); print('device:', select_device('auto'))
```

```python
import cv2, matplotlib.pyplot as plt, numpy as np
from src.preprocessing import build_augmentation_pipeline, letterbox
from src.data import list_images

img = cv2.imread(str(list_images(root / 'data/processed/sh17/images/train')[0]))
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
aug = build_augmentation_pipeline(640, train=True)
```

```python
fig, axes = plt.subplots(2, 3, figsize=(15, 8))
axes[0,0].imshow(img); axes[0,0].set_title('original'); axes[0,0].axis('off')
for ax in axes.ravel()[1:]:
    out = aug(image=img, bboxes=[], class_labels=[])['image'] if aug else letterbox(img)
    ax.imshow(out); ax.set_title('augmented'); ax.axis('off')
plt.show()
```

`hsv_*`, `mosaic`, `fliplr`, con `degrees=0` e `flipud=0` per il dominio del cantiere in posizione eretta.
